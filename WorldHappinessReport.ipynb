```python
corruption = df.groupby('Region')[['Perceptions_of_corruption']].mean()
corruption
```

```python
plt.title('Perception of Corruption in Various Regions')
plt.xlabel('Region')
plt.ylabel('Corruption Index')
plt.xticks(rotation=30, ha='right')
plt.bar(corruption.index, corruption.Perceptions_of_corruption)
```

```python
top_10 = df.head(10)
top_10
```

```python
bottom_10 = df.tail(10)
bottom_10
```

```python
df.columns
```

```python
x = top_10
y = bottom_10
plt.title('Life Expectancy of Top 10 Happiest Countries')
plt.xticks(rotation=30, ha='right')
sns.barplot(x=top_10.Country, y=top_10.Healthy_life_expectancy)
```

```python
plt.title('Life Expectancy of Bottom 10 Happiest Countries')
plt.xticks(rotation=30, ha='right')
sns.barplot(x=bottom_10.Country, y=bottom_10.Healthy_life_expectancy)
```

```python
# Happiness score vs Freedom to Make Life Choices
plt.rcParams['figure.figsize']=(15,7)
sns.scatterplot(x=df.Freedom_to_make_life_choices, y=df.Ladder_score, hue=df.Region, s=100)
plt.legend(loc = 'upper left', fontsize = '12')
plt.xlabel('Freedom to Make Life Choices')
plt.ylabel('Happiness Score')
```

```python
# Corruption vs Happiness
plt.rcParams['figure.figsize']=(15,7)
sns.scatterplot(x=df.Ladder_score, y=df.Perceptions_of_corruption, hue=df.Region, s=100)
plt.legend(loc = 'lower left', fontsize = '12')
plt.xlabel('Happiness Score')
plt.ylabel('Corruption')
```
